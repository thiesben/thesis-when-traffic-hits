## Reading in Trip files
At first, we read in trip files for further analysis. We only focus on a subset of columns, as the others are either deprecated or not relevant for our analysis

```python
# Imports

import contextily as ctx
import geopandas as gpd
import matplotlib.pyplot as plt
import multiprocessing as mp
import numpy as np
import os
import pandas as pd
from shapely.geometry import LineString

from config import rpath
os.chdir(rpath)

# import pyproj
# from config import projpath
# pyproj.datadir.set_data_dir(projpath)
```

```python
path = "data/Berlin_2017/data/trips"
files = os.listdir(path)
colnames = ["TripID","DeviceID","ProviderID","Mode","StartDate","StartWday","EndDate","EndWDay","StartLocLat",
            "StartLocLon", "EndLocLat","EndLocLon","IsStartHome","IsEndHome", "GeospatialType","ProviderType",
            "ProviderDrivingProfile","VehicleWeightClass","ProbeSourceType","OriginZoneName","DestinationZoneName",
            "MultipleZones","MultipleCorridors","EndpointType","TripMeanSpeedKph","TripMaxSpeedKph","TripDistanceMeters",
            "MovementType","OriginCensusBlockGroup","DestinationCensucBlockGroup"]
keepcols = ["TripID","DeviceID","StartDate","EndDate","StartLocLat","StartLocLon", "EndLocLat","EndLocLon", "ProviderType",
            "ProviderDrivingProfile","VehicleWeightClass","TripMeanSpeedKph","TripMaxSpeedKph","TripDistanceMeters",
            "MovementType"]
```

```python
for i, file in enumerate(files):
    if i== 0:
        df = pd.read_csv(os.path.join(path,file), names = colnames)
        df = df[keepcols]
    else:
        df = df.append(pd.read_csv(os.path.join(path,file), names=colnames)[keepcols])
    print("Read file", i+1, "of", len(files))
```

In the following, we get the total number of unique trips (*N*), draw a random sample and save it as npy file for further usage.

```python
N = len(set(df.TripID.tolist()))
print(N)
print(int(np.round(N*0.02,0)))
```

```python
trip_sample = df["TripID"].sample(n=int(np.round(N*0.02,0)), random_state=1).tolist()
np.save("tripsample.npy", np.array(trip_sample))
```

## Reading in Waypoint files
Here, we first read in all the waypoints and filter for those which belong to a trip out of the trip id sample, that is created in the trips notebook.

```python
def read_waypoint_file(file):
    df = pd.read_csv(file, names = colnames)
    df = df[keepcols]
    df = df[df['TripID'].isin(tripids)]
    print("Read file", file)
    print("File contained", len(df), "matching entries.")
    return df
```

```python
path = "data/Berlin_2017/data/waypoints"
files = []
for file in os.listdir(path):
    if file.endswith("edges.csv"):
        continue
    files.append(os.path.join(path,file))
colnames = ['TripID','WaypointSequence','CaptureDate','lat','lon','SegmentID', 
            'ZoneName','Frc','DeviceID','RawSpeed','RawSpeedMetric']
keepcols = ['TripID','WaypointSequence','CaptureDate','lat','lon','DeviceID','RawSpeed']
tripids = np.load("data/plots/tripsample.npy")
```

```python
pool = mp.Pool(processes=15)
df_list = pool.map(read_waypoint_file, files)
pool.close()
```

```python
df = df_list.pop(0)
for i, d in enumerate(df_list):
    df = df.append(d)
    print(i+1, "files processed")
del df_list
```

```python
df.to_csv("data/plots/waypointsample.csv")
df.head()
```

## Creating the GeoDataFrame
Next, we create an actual GeoDataFrame, set the crs to 4326 (Lat/Lon) and convert it to UTM 33N to have it metric. Then, we save it so that we don't have to do it all again.

```python
df = pd.read_csv("data/plots/waypointsample.csv")
print("Waypoint-N:", len(df))
# Only include roughly Berlin since there are still 31 Mio waypoints in df
df = df[(df['lat'] >= 52.343774) & (df['lat'] <= 52.703684) & (df['lon'] >= 13.083344) & (df['lon'] <= 13.842607)] 
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat))
del df # for space
gdf = gdf.set_crs(4326)
```

```python
parts = np.array_split(gdf, 15)
del gdf
new_parts = []
```

```python
for i in range(len(parts)):
    new_parts.append(parts[0].to_crs(25833))
    print(i+1, "parts processed")
    parts.pop(0)
gdf = new_parts.pop(0)
for i, d in enumerate(new_parts):
    gdf = gdf.append(d)
    print(i+1, "parts processed")
del new_parts
gdf["X_258333"] = gdf.geometry.x
gdf["Y_258333"] = gdf.geometry.y
```

```python
# Save as csv b/c Shapefile doesn't really make sense b/c of its size and handling
gdf.drop(columns=['geometry']).to_csv("data/plots/waypointsample-25833.csv")
```

Here, we create lines between points "as the crow flies" and save the resulting Data Frame

```python
# Remove trips with only one waypoint
gdf['count'] = gdf.groupby('TripID')['TripID'].transform('count')
gdf = gdf[gdf["count"]>1]
```

```python
gdf_lines = gdf.groupby(['TripID'])['geometry'].apply(lambda x: LineString(x.tolist()))
gdf_lines = gdf_lines.set_crs(25833)
gdf_lines = gpd.GeoDataFrame(gdf_lines)
gdf_lines.to_file("data/plots/waypointsample-lines.shp")
```

# Plotting
Here, we create plots. We assume, that `gdf_lines` and `gdf` are not in the environment. First, we want to replicate the plot from Bazzani et al. (2010). For this, we create the `RawSpeed` variable and then plot the network in a given bounding box.

```python
# Load data
gdf = pd.read_csv("data/plots/waypointsample-25833.csv")
gdf = gpd.GeoDataFrame(gdf, geometry=gpd.points_from_xy(gdf.lon, gdf.lat)).set_crs(4326)
```

```python
conditions = [
    (gdf['RawSpeed'] <= 30),
    (gdf['RawSpeed'] > 30) & (gdf['RawSpeed'] <= 60),
    (gdf['RawSpeed'] > 60)
    ]
values = ['red','yellow','green']
gdf['speedcolor'] = np.select(conditions, values)
```

```python
gdf_ber_small = gdf[(gdf['lat'] >= 52.509493) & (gdf['lat'] <= 52.556567) &
                        (gdf['lon'] >= 13.274437) & (gdf['lon'] <= 13.456668)] 
gdf_ber_small = gdf[(gdf['lat'] >= 52.504064) & (gdf['lat'] <= 52.534955) &
                        (gdf['lon'] >= 13.270889) & (gdf['lon'] <= 13.324895)] 
print(len(gdf_ber_small))
```

```python
gdf_small = gdf[(gdf['lat'] >= 52.499560) & (gdf['lat'] <= 52.555780) &
                        (gdf['lon'] >= 13.274731) & (gdf['lon'] <= 13.463213)] 
df[(df['lat'] >= 52.343774) & (df['lat'] <= 52.703684) & (df['lon'] >= 13.083344) & (df['lon'] <= 13.842607)] 
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
gdf_small.iloc[:,:].plot(color=gdf_small['speedcolor'], markersize=0.05, ax=ax)
ax.set_facecolor('black')
plt.margins(x=0,y=0)
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
gdf_ber_small.iloc[:,:].plot(color=gdf_ber_small['speedcolor'], markersize=0.05, ax=ax)
ax.set_facecolor('black')
plt.margins(x=0,y=0)
plt.show()
```
