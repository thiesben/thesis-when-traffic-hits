# Cross-Validation of ZIP and ZINB

```python
upper_thresh = 20
predictors = ["log_flow_n", "maxSpeed"]
target = "n_accidents"

# Get data
all_data = edges_accidents[edges_accidents.n_accidents <= upper_thresh]
all_data = all_data.sample(frac=1) # shuffle data randomly
```

```python
# Prepare folds
N_folds = 10
limits = np.linspace(0, len(all_data)+1, N_folds+1, dtype=int)
folds = [(limits[i], limits[i+1]) for i in range(len(limits)-1)]
```

```python
zinb_rmses = []
zip_rmses = []
zip_maes = []
zinb_maes = []

for fold in folds:
    
    foldbool = np.zeros(len(all_data), dtype=bool)
    foldbool[fold[0]:fold[1]] = True # indexer for test data
    
    
    
    test = all_data[foldbool]
    train = all_data[~foldbool]

    zip_training_results = ZeroInflatedPoisson(endog = train[target], 
                                               exog = add_constant(train[predictors]),
                                               exog_infl = add_constant(train[predictors]), 
                                               inflation = 'logit').fit(maxiter=100)

    zip_predictions = zip_training_results.predict(add_constant(test[predictors]),
                                                   exog_infl=add_constant(test[predictors]))
    zip_predicted_counts = np.round(zip_predictions)

    zip_rmses.append(np.sqrt(np.mean(np.power(np.subtract(zip_predicted_counts,y_test),2))))
    zip_maes.append(np.mean(np.absolute(np.subtract(zip_predicted_counts,y_test))))   
    
    zinb_training_results = ZeroInflatedNegativeBinomialP(endog = train[target], 
                                                          exog = add_constant(train[predictors]),
                                                          exog_infl = add_constant(train[predictors]), 
                                                          inflation = 'logit').fit(maxiter=100)

    zinb_predictions = zinb_training_results.predict(add_constant(test[predictors]),
                                                     exog_infl=add_constant(test[predictors]))
    zinb_predicted_counts = np.round(zinb_predictions)
    
    zinb_rmses.append(np.sqrt(np.mean(np.power(np.subtract(zinb_predicted_counts,y_test),2))))
    zinb_maes.append(np.mean(np.absolute(np.subtract(zinb_predicted_counts,y_test))))                                              
```

```python
import matplotlib.pyplot as plt

x = np.array(range(len(folds)))
 
# colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
# print(colors)

fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(x, zinb_rmses, label="ZINB")
ax[0].plot(x, zip_rmses, label="ZIP")
ax[0].plot(x, np.repeat(np.mean(zinb_rmses),len(x)), color="#1f77b4", linestyle="-.", label="ZINB mean")
ax[0].plot(x, np.repeat(np.mean(zip_rmses),len(x)), color="#ff7f0e", linestyle="-.", label="ZIP mean")
ax[0].set_xticks(x)
ax[0].set_xticklabels(x+1)
ax[0].set_xlabel("Fold", fontsize=12)
ax[0].set_ylabel("RMSE", fontsize=12)
ax[0].legend()


ax[1].plot(x, zinb_maes, label="ZINB")
ax[1].plot(x, zip_maes, label="ZIP")
ax[1].plot(x, np.repeat(np.mean(zinb_maes),len(x)), color="#1f77b4", linestyle="-.", label="ZINB mean")
ax[1].plot(x, np.repeat(np.mean(zip_maes),len(x)), color="#ff7f0e", linestyle="-.", label="ZIP mean")
ax[1].set_xticks(x)
ax[1].set_xticklabels(x+1)
ax[1].set_xlabel("Fold", fontsize=12)
ax[1].set_ylabel("MAE", fontsize=12)
ax[1].legend(loc="upper left")
plt.show()


print("zip RMSE:", np.mean(zip_rmses))
print("zinb RMSE:", np.mean(zinb_rmses))
print("zip MAE:", np.mean(zip_maes))
print("zinb MAE:", np.mean(zinb_maes))


print("mean of n_accidents (w/o 0s):", np.mean(all_data[all_data.n_accidents>0][target]))
print("variance of n_accidents (w/o 0s):", np.std(all_data[all_data.n_accidents>0][target])**2)

# TODO: ADD MEAN VALUE TO THE GRAPH
```

Next, we can build the hinge classifier. We use a common logistic regression classifier.

```python
#threshs = np.arange(0.1,1,.05)
threshs=[.6]
for thresh in threshs:
    print(thresh)
    # Specify train / test data
    log_X_train = df_train[["log_flow_n", "log_flow_n_2"]]
    log_y_train = df_train["has_accident"]
    log_X_test = df_test[["log_flow_n", "log_flow_n_2"]]
    log_y_test = df_test["has_accident"]

    # Create / fit model
    logmodel = LogisticRegression(solver='liblinear', random_state=0)
    logmodel.fit(log_X_train, log_y_train)

    # Test model
    log_yhat_proba = logmodel.predict_proba(log_X_test)
    log_yhat = [int(a[1]>=thresh) for a in log_yhat_proba]
    print("Prediction accuracy:", np.mean(log_y_test == log_yhat))

    # Predict the y values for the given x
    log_yhat_train = logmodel.predict_proba(log_X_train)
    log_yhat_train_plot = [a[1] for a in log_yhat_train]

    log_yhat_train_class = [int(a[1]>=thresh) for a in log_yhat_train]
    
    # Plot the outcome
    fig, ax = plt.subplots(1,1)
    ax.scatter(log_X_train["log_flow_n"], log_yhat_train_plot,color="orange")
    ax.scatter(log_X_train["log_flow_n"], log_yhat_train_class)

    # Print accuracy of the model (72%)
    print("Train Accuracy:",np.mean(log_yhat_train_class==log_y_train))

    plt.show()
```

```python
def plot_linear_model(res_object, x_data, x_var, y_data):
    # Plot the linear model
    plt.plot(x_data[x_var], y_data, 'o', label='original data')
    
    ranges = []
    
    for var in res_object.params[1:].index:
        ranges.append(np.linspace(x_data[var].min(), x_data[var].max(),100))
    xrange = np.linspace(x_data[x_var].min(), x_data[x_var].max(),100)

    yhat_values = np.zeros_like(ranges[0]) + res_object.params[0]
    
    for i, coef in enumerate(res_object.params[1:]):
        
        yhat_values = yhat_values + coef*ranges[i] 
    
    plt.plot(xrange,
             yhat_values,
             'r-', label='fitted line', linewidth=3)
    plt.legend()
    plt.show()

plot_linear_model(res, lin_X_train, "log_flow_n", lin_y_train)
```
